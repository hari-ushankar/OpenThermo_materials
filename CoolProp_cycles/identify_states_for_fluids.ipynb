```python
import CoolProp.CoolProp as CP
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import TextBox, Button

# Define the substance and range of temperatures and pressures
substance = 'Water'
T_min = 100  # Minimum temperature in Kelvin (0 degrees Celsius)
T_max = 447.096  # Critical temperature iaan Kelvin
P_min = 611.657  # Saturation pressure at the triple point in Pa
P_max = 220.04e5  # Maximum pressure in Pa

# Create an array of temperatures and pressures
T_vals = np.linspace(T_min, T_max, 100)
P_vals = np.linspace(P_min, P_max, 1000)

# Create a P-T phase diagram
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), dpi=400)

for P in P_vals:
    # Calculate temperature values for the given pressure at saturated liquid and vapor states
    T_liquid = CP.PropsSI('T', 'P', P, 'Q', 0, 'Water')
    T_vapor = CP.PropsSI('T', 'P', P, 'Q', 1, 'Water')

    # Plot the P-T diagram
    ax1.plot([T_liquid - 273.15, T_vapor - 273.15], [P / 1e3, P / 1e3], '*b')

# Set plot attributes for P-T diagram
ax1.set_xlabel('Temperature (C)')
ax1.set_ylabel('Pressure (kPa)')
ax1.set_title('Pressure-Temperature (P-T) Phase Diagram for Water')
ax1.grid(True, ls='--')

# Define the substance and range of specific volumes and pressures
P_min = 611.657  # Saturation pressure at the triple point in Pa
P_max = 22.004e6  # Maximum pressure in Pa

# Create an array of specific volumes and pressures
#V_vals = np.logspace(np.log10(V_min), np.log10(V_max), 100)
P_vals = np.linspace(P_min, P_max, 1000)

# Create a P-V diagram
for P in P_vals:
    # Calculate specific volume values for the given pressure at saturated liquid and vapor states
    V_liquid = 1 / CP.PropsSI('D', 'P', P, 'Q', 0, 'Water')
    V_vapor = 1 / CP.PropsSI('D', 'P', P, 'Q', 1, 'Water')

    # Plot the P-V diagram
    ax2.semilogx([V_liquid, V_vapor], [P / 1e3, P / 1e3], '*b')

# Set plot attributes for P-V diagram
ax2.set_xlabel('Specific Volume (m/kg)')
ax2.set_ylabel('Pressure (kPa)')
ax2.set_title('Pressure-Volume (P-V) Diagram for Water (Semilog Scale)')
ax2.grid(True, ls='--')

plt.tight_layout()
plt.show()
```

```python
import CoolProp.CoolProp as CP

# Define fluid and state points
fluid = 'Water'
P1 = 10e3  # Pump inlet pressure in Pa
P2 = 2e6   # Boiler inlet pressure in Pa
P3 = 10e3  # Turbine inlet pressure in Pa
P4 = 10e3  # Condenser inlet pressure in Pa

# Isentropic efficiencies of pump and turbine
efficiency_pump = 0.75
efficiency_turbine = 0.85

# Calculate state properties for each step of the cycle
# State 1 (Pump inlet)
h1 = CP.PropsSI('H', 'P', P1, 'Q', 0, fluid)
s1 = CP.PropsSI('S', 'P', P1, 'Q', 0, fluid)

# State 2 (Boiler inlet)
h2s = CP.PropsSI('H', 'P', P2, 'S', s1, fluid)  # Isentropic enthalpy at state 2
h2 = h1 + efficiency_pump * (h2s - h1)

# State 3 (Turbine inlet)
h3s = CP.PropsSI('H', 'P', P3, 'S', s1, fluid)  # Isentropic enthalpy at state 3
h3 = h2 - efficiency_turbine * (h2 - h3s)

# State 4 (Condenser inlet)
h4 = CP.PropsSI('H', 'P', P4, 'Q', 0, fluid)
s4 = CP.PropsSI('S', 'P', P4, 'Q', 0, fluid)

# Calculate cycle parameters
Q_in = h2 - h3  # Heat added in the boiler
Q_out = h1 - h4  # Heat rejected in the condenser
W_net = Q_in - Q_out  # Net work output

# Print the results
print("State 1 (Pump Inlet):")
print(f"Enthalpy (h1): {h1} J/kg")
print(f"Entropy (s1): {s1} J/(kg-K)")

print("\nState 2 (Boiler Inlet):")
print(f"Enthalpy (h2): {h2} J/kg")

print("\nState 3 (Turbine Inlet):")
print(f"Enthalpy (h3): {h3} J/kg")

print("\nState 4 (Condenser Inlet):")
print(f"Enthalpy (h4): {h4} J/kg")
print(f"Entropy (s4): {s4} J/(kg-K)")

print("\nCycle Parameters:")
print(f"Heat added (Q_in): {Q_in} J/kg")
print(f"Heat rejected (Q_out): {Q_out} J/kg")
print(f"Net work output (W_net): {W_net} J/kg")
```

## Calculate work done with Rankine cycle on a P-v diagram (steam power plants)

```python
import CoolProp.CoolProp as CP
import numpy as np
import matplotlib.pyplot as plt

# Define fluid and state points
fluid = 'Water'
P1 = 10e3  # Pump inlet pressure in Pa
P2 = 2e6   # Boiler inlet pressure in Pa
P3 = 10e3  # Turbine inlet pressure in Pa
P4 = 10e3  # Condenser inlet pressure in Pa

# Isentropic efficiencies of pump and turbine
efficiency_pump = 0.75
efficiency_turbine = 0.85

# Create an array of specific volumes and pressures for the P-V diagram
V_vals = np.linspace(1e-4, 1.5, 1000)  # Specific volume in m/kg
P_vals = np.linspace(1e3, 3e6, 1000)   # Pressure in Pa

# Calculate state properties for each step of the cycle
# State 1 (Pump inlet)
h1 = CP.PropsSI('H', 'P', P1, 'Q', 0, fluid)
s1 = CP.PropsSI('S', 'P', P1, 'Q', 0, fluid)

# State 2 (Boiler inlet)
h2s = CP.PropsSI('H', 'P', P2, 'S', s1, fluid)  # Isentropic enthalpy at state 2
h2 = h1 + efficiency_pump * (h2s - h1)

# State 3 (Turbine inlet)
h3s = CP.PropsSI('H', 'P', P3, 'S', s1, fluid)  # Isentropic enthalpy at state 3
h3 = h2 - efficiency_turbine * (h2 - h3s)

# State 4 (Condenser inlet)
h4 = CP.PropsSI('H', 'P', P4, 'Q', 0, fluid)
s4 = CP.PropsSI('S', 'P', P4, 'Q', 0, fluid)

# Calculate specific volumes for each state point
V1 = 1 / CP.PropsSI('D', 'P', P1, 'H', h1, fluid)
V2 = 1 / CP.PropsSI('D', 'P', P2, 'H', h2, fluid)
V3 = 1 / CP.PropsSI('D', 'P', P3, 'H', h3, fluid)
V4 = 1 / CP.PropsSI('D', 'P', P4, 'H', h4, fluid)

# Plot the P-V diagram and the process cycle
plt.figure(dpi=100)
plt.plot([V1, V2, V3, V4, V1], [P1 / 1e3, P2 / 1e3, P3 / 1e3, P4 / 1e3, P1 / 1e3], 'o-b')
plt.xlabel('Specific Volume (m/kg)')
plt.ylabel('Pressure (kPa)')
plt.title('Rankine Cycle on P-V Diagram')
plt.grid(True, ls='--')
plt.show()
```

## Brayton cycle caculation on a P-v cycle (idealized model for gas turbine engines)

```python
import CoolProp.CoolProp as CP
import numpy as np
import matplotlib.pyplot as plt

# Define fluid and state points
fluid = 'Air'
P1 = 1e5  # Compressor inlet pressure in Pa
P2 = 2e5  # Combustion chamber inlet pressure in Pa
P3 = 2e5  # Turbine inlet pressure in Pa
P4 = 1e5  # Exhaust pressure in Pa

# Isentropic efficiencies of compressor and turbine
efficiency_comp = 0.5
efficiency_turbine = 0.5

# Create an array of specific volumes and pressures for the P-V diagram
V_vals = np.linspace(1e-4, 2e-3, 1000)  # Specific volume in m/kg
P_vals = np.linspace(1e5, 3e5, 1000)  # Pressure in Pa

# Calculate state properties for each step of the cycle
# State 1 (Compressor inlet)
h1 = CP.PropsSI('H', 'P', P1, 'Q', 1, fluid)
s1 = CP.PropsSI('S', 'P', P1, 'Q', 1, fluid)

# State 2 (Combustion chamber inlet)
h2s = CP.PropsSI('H', 'P', P2, 'S', s1, fluid)  # Isentropic enthalpy at state 2
h2 = h1 - efficiency_comp * (h1 - h2s)

# State 3 (Turbine inlet)
h3s = CP.PropsSI('H', 'P', P3, 'S', s1, fluid)  # Isentropic enthalpy at state 3
h3 = h2 - efficiency_turbine * (h2 - h3s)

# State 4 (Exhaust)
h4 = CP.PropsSI('H', 'P', P4, 'Q', 1, fluid)
s4 = CP.PropsSI('S', 'P', P4, 'Q', 1, fluid)

# Calculate specific volumes for each state point
V1 = 1 / CP.PropsSI('D', 'P', P1, 'H', h1, fluid)
V2 = 1 / CP.PropsSI('D', 'P', P2, 'H', h2, fluid)
V3 = 1 / CP.PropsSI('D', 'P', P3, 'H', h3, fluid)
V4 = 1 / CP.PropsSI('D', 'P', P4, 'H', h4, fluid)

# Plot the P-V diagram and the Brayton cycle
plt.figure(dpi=100)
plt.plot([V1, V2, V3, V4, V1], [P1 / 1e3, P2 / 1e3, P3 / 1e3, P4 / 1e3, P1 / 1e3], '-b')
plt.xlabel('Specific Volume (m/kg)')
plt.ylabel('Pressure (kPa)')
plt.title('Brayton Cycle on P-V Diagram')
plt.grid(True, ls='--')
plt.show()
```

## Otto cycle for P-v cycle (idealized spark-ignition internal combustion engines)

```python
import CoolProp.CoolProp as CP
import numpy as np
import matplotlib.pyplot as plt

# Define the working fluid
fluid = 'Air'

# Initial conditions (State 1)
P1 = 101325  # Initial pressure (Pa)
T1 = 300  # Initial temperature (K)

# Compression ratio (V2/V1)
compression_ratio = 10

# Calculate state properties for State 2 (End of Compression)
V1 = 1 / CP.PropsSI('D', 'P', P1, 'T', T1, fluid)
V2 = V1 / compression_ratio
P2 = P1 * compression_ratio

# Calculate state properties for State 3 (End of Heat Addition)
T3 = 2200  # Temperature at the end of heat addition (K)
P3 = P2
V3 = CP.PropsSI('V', 'P', P3, 'T', T3, fluid)

# Calculate state properties for State 4 (End of Expansion)
P4 = P1
V4 = V3 / compression_ratio

# Calculate specific heat capacities at constant pressure and constant volume
cp = CP.PropsSI('C', 'P', P1, 'T', T1, fluid)
cv = CP.PropsSI('CVMASS', 'P', P1, 'T', T1, fluid)

# Calculate the isentropic exponent (gamma)
gamma = cp / cv

# Create arrays of specific volumes for each process
V_vals_comp = np.linspace(V1, V2, 100)
V_vals_heat = np.linspace(V2, V3, 100)
V_vals_exp = np.linspace(V3, V4, 100)

# Calculate pressures for each process
P_vals_comp = [P1 * (V2 / V) ** gamma for V in V_vals_comp]
P_vals_heat = [P3 for _ in V_vals_heat]
P_vals_exp = [P4 * (V4 / V) ** gamma for V in V_vals_exp]

# Plot the P-V diagram
plt.figure(figsize=(8, 6))
plt.plot(V_vals_comp, P_vals_comp, 'b-', label='Adiabatic Compression')
plt.plot(V_vals_heat, P_vals_heat, 'r-', label='Constant Volume Heat Addition')
plt.plot(V_vals_exp, P_vals_exp, 'g-', label='Adiabatic Expansion')
plt.scatter([V1, V2, V3, V4], [P1, P2, P3, P4], color='black', label='States')
plt.xlabel('Specific Volume (m/kg)')
plt.ylabel('Pressure (Pa)')
plt.title('P-V Diagram for Otto Cycle')
plt.grid(True, ls='--')
plt.legend()
plt.show()
```

## Simple thermodynamic cycle calculations inside Coolprop; assign states for each cycle

```python
import CoolProp
from CoolProp.Plots import PropertyPlot
from CoolProp.Plots import SimpleCompressionCycle
pp = PropertyPlot('HEOS::R134a', 'PH', unit_system='EUR')
pp.calc_isolines(CoolProp.iQ, num=11)
cycle = SimpleCompressionCycle('HEOS::R134a', 'PH', unit_system='EUR')
T0 = 280
pp.state.update(CoolProp.QT_INPUTS,0.0,T0-10)
p0 = pp.state.keyed_output(CoolProp.iP)
T2 = 310
pp.state.update(CoolProp.QT_INPUTS,1.0,T2+15)
p2 = pp.state.keyed_output(CoolProp.iP)
pp.calc_isolines(CoolProp.iT, [T0-273.15,T2-273.15], num=2)
cycle.simple_solve(T0, p0, T2, p2, 0.7, SI=True)
cycle.steps = 1000
sc = cycle.get_state_changes()
pp.draw_process(sc)
import matplotlib.pyplot as plt
plt.close(cycle.figure)
pp.show()
```
